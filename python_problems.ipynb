Review: Dictionaries are key and value pairs. Keys are usually for something thats easy to remember, which is used to get us the value, something that is normally harder to remember. In this example, we will use the food as the key, and price as the value.

```
food_dict = {

'hotdog' : 1.50,
'acai bowl' : 4.00,
'pizza slice' : 2.00

}
```
recall that if we want to get the price of the hotdog, we simply do

```
hotdog_price = food_dict['hotdog']
print(hotdog_price) # will print out 1.50
```

to change the price of the hotdog, we do something similar

```
food_dict['hotdog'] = 2.00
hotdog_price = food_dict['hotdog']
print(hotdog_price) # will now print out 2.00, our new price
```

````
# Starter code for the Dolphin class
######################################
# DO NOT MODIFY.
######################################
class Dolphin:
  
  # Recall that the __init__() function is a python built in function for classes
  # that defines our class variables. The variables defined here will be specific
  # to Dolphins.
  def __init__(self):
    
    self.name = None    # dolphins have different names, so we use a placeholder None here
    self.canSwim = True # all dolphins can swim, I hope, so we set a boolean value here as True
  
  # this method makes the dolphin growl
  def growl(self):
    print("GRRRRRR")
````

````
# Recall that we can create and set the name of the dolphin like so

dolphin_1 = Dolphin()     # tells the variable dolphin_1 that its now a dolphin
dolphin_1.name = 'Bottle' # sets dolphin_1's name to Bottle
print(dolphin_1.name)     # prints out dolphin_1's name
dolphin_1.growl()         # tells dolphin_1 to growl
````
